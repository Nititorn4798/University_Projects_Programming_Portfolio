**ข้อ19**

```python
plt.style.use('seaborn-v0_8')

y = [10, 15, 15, 20, 10]
x = np.arange(1, 6)

bw = 0.25

plt.bar(x, y, color='g', width=bw, label='สินค้า #1')

y = [15,25,10,15,20]
plt.bar(x + bw, y , color='lightblue', width=bw, label='สินค้า #2')

y = [10,20,15,10,15]
plt.bar(x + (2 * bw), y, hatch='..', color='lightgray',
        width=bw, label='สินค้า #3')

plt.xticks(x + bw, [2016,2017,2018,2019,2020])

plt.rcParams['font.family'] = 'Sarabun'
plt.rcParams['font.size'] = 14
plt.legend(loc='best')

plt.show()
```
